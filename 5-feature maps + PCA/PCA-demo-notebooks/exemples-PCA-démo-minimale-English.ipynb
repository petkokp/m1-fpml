```python
## number fo dimensions to reduce the data to.
ACP_analysis = 60
```

```python
## PCA: quick demo
## no labels required !
import numpy as np
import matplotlib.pyplot as plt
X = np.load('/home/flandes/data/fashion-mnist-reshaped.npz_FILES/train_images.npy')
import sklearn.decomposition

## define an object of the class  decomposition.PCA
classe_PCA = sklearn.decomposition.PCA(n_components=ACP_analysis, random_state=42)
classe_PCA.fit(X)  ## permet de toruver autoomatiquement la valeur de la matrice de projection P
transformationMatrix = classe_PCA.components_.transpose() ## la matrice de projection P

## reduce the data X to an array of 30-dimensional vectors (30 pixels "images")
X_transformed = classe_PCA.transform(X) # automatically
# X_transformed = np.dot(X - np.mean(X, axis=0), transformationMatrix) ## by hand

## go back to the original space's dimensions (for visualization purposes)
X_reExpanded = classe_PCA.inverse_transform(X_transformed)
```

```python
## (for visualization purposes)
for imgNum in[0,1,4,742]:
    f, axarr = plt.subplots(1,2, figsize=(12,12))
    img = X[imgNum].reshape(28,28)
    axarr[0].imshow(img, plt.cm.gray)
    
    img = X_reExpanded[imgNum].reshape(28,28)
    axarr[1].imshow(img, plt.cm.gray)
```
